```python
import pandas as pd
import numpy as np
```

# Load Data

```python
df=pd.read_csv('Global Missing Migrants Dataset.csv')
```

```python
df.head()
```

```python
df.shape
```

# Clean Data

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Region of Origin']=df['Region of Origin'].fillna(df['Region of Origin'].mode()[0])
df['Country of Origin']=df['Country of Origin'].fillna(df['Country of Origin'].mode()[0])
df['Information Source']=df['Information Source'].fillna(df['Information Source'].mode()[0])
df['Number of Dead']=df['Number of Dead'].fillna(df['Number of Dead'].mean())
df['Migration route']=df['Migration route'].fillna(df['Migration route'].mode()[0])
```

```python
df['Number of Dead']=df['Number of Dead'].astype(int)
```

```python
df['Number of Dead']
```

```python
df.isnull().sum()
```

```python
df['Coordinates']=df['Coordinates'].fillna(df['Coordinates'].mode()[0])
df['UNSD Geographical Grouping']=df['UNSD Geographical Grouping'].fillna(df['UNSD Geographical Grouping'].mode()[0])
```

```python
df.columns
```

```python
df['Minimum Estimated Number of Missing'] = df['Minimum Estimated Number of Missing'].apply(lambda x: np.nan if x < 0 else x)
```

```python
df['Minimum Estimated Number of Missing'].min()
```

```python
df['Minimum Estimated Number of Missing']=df['Minimum Estimated Number of Missing'].fillna(df['Minimum Estimated Number of Missing'].mean())
df['Minimum Estimated Number of Missing']=df['Minimum Estimated Number of Missing'].astype(int)
```

```python
df['Number of Survivors'] = df['Number of Survivors'].apply(lambda x: np.nan if x < 0 else x)
df['Number of Survivors'].min()
```

```python
df['Number of Survivors']=df['Number of Survivors'].fillna(df['Number of Survivors'].mean())
df['Number of Survivors']=df['Number of Survivors'].astype(int)
```

```python
df.to_csv("Global Missing Migrants Dataset.csv", index=False)
```
